```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

stars = ['O5','B5','A5','F5','G5','K5','M5']
temp_stars = [54000, 15200, 8310, 6700, 5660, 4400, 3200]
temp = np.array(temp_stars)
max_wavelength = (2.898*10**(-3))/(temp*10)
plt.scatter(temp, max_wavelength)
plt.xlabel("Temperature(K)")
plt.ylabel("Wavelength(m)")
for x, label in enumerate(stars):
    plt.annotate(label, (temp_stars[x], max_wavelength[x]))
           
```

A note about units: recall from PHYS 208/245 that 1 Volt = 1 Joule/Coulomb. Electrons and protons give us a fundamental unit of electric charge, or elementary charge, $|q| = e = 1.602 \times 10^{−19}$ C. We can also calculate a fundamental unit of energy by calculating the change in kinetic energy of the elementary charge $e$ accelerated through an electric potential of 1 Volt. This new unit of energy is called the electron-Volt or eV. 1 eV = $1.602 \times 10^{−19}$ J.

    <li>Assume you are trying to verify the experimental results that led to Einstein's successful explanation of the photoelectric effect. You have two emitters or "photocathodes"; one is made of aluminum and the other is made of gold. On <strong>one set of axes</strong>, plot the <em>maximum</em> photoelectron kinetic energy in eV that you could measure as a function of frequency $\nu$ from <strong>both photocathodes</strong>. Make sure your plot shows the frequency cutoffs below which <em>no</em> photoelectrons will be generated. Label both axes and put a legend on your plot so a reader with no physics experience will know which curve belongs to which metal.
```python
c = constants.c * (10**9)
f = np.linspace(-(c/1000),c/100, num = 100, endpoint = True)
energy = (4.14*(10**(-15)))*f
gold = energy - 5.1
al = energy - 4.08
fig, ax = plt.subplots()
ax.plot(f, al, color = 'b', label = "KE Aluminum")
ax.plot(f, gold, color = 'purple', label = "KE Gold")
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Max Photoelectron KE (eV)")
#Aluminum: 4.08 eV = (4.14*10^(-15))(f) f = 9.85*10^(14) Hz
# Gold: 5.1 eV = (4.14*10^(-15))(f) f = 1.23*10(15) Hz
ax.axvline(9.85*(10**14), color = 'g', label = "Aluminum Cutoff")
ax.axvline(1.23*(10**15), color = 'r', label = "Gold Cutoff")
plt.show()
```
